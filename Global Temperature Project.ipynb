# Global Temperature
In this project, you will analyze local and global temperature data and compare the temperature trends where you live to overall global temperature trends.My goal is to create a visualization and prepare a write up describing the similarities and differences between global temperature trends and temperature trends in the closest big city to where I live.

After extracting data with queries in SQL, I used the excel to check and clean data. In Excel I utilized =average()  function to find 10 years moving average.
Finally, I used Jupyter Notebook to make visualization in Phyton.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("Temprature.csv")
```

```python
df.head(15)
```

```python
#after all, we see that there are some NaN values on the data, 
#for this reason we use dropna function to delete these rows from data. 
df_t=df.dropna()
df_t
```

```python
start = df_t.min(axis=0)['cyear']
end = df_t.max(axis=0)['cyear']
```

```python
# At last we make visualization to compare global average temperature and average temperature of Berlin
plt.figure(figsize=(10, 6))
plt.title('Global vs Berlin Temperature Change 1760 - 2013')
plt.plot( df_t['cyear'], df_t['city_mavg_temp'], label='Berlin')
plt.plot( df_t['cyear'], df_t['mavg_temp'],label='Global')
plt.xlabel('Year')
plt.ylabel('Temperature(C)')
plt.xticks(np.arange(start, end,25))
plt.legend(loc=4)
plt.show()
```
